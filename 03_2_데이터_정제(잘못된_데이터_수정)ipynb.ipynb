># 누락된 값 처리하기

- 누락된 값을 다른 값으로 바꾼다.

>>## 누락된 값 개수 확인하기: `isna()`
- `isna()` 메서드는 각 행이 비어있는지를 나타내는 불리언 배열을 반환
- `sum()` 메서드를 이어서 호출하면 불리언 배열의 True 개수로 비어있는 행 개수를 얻을 수 있음

```python
ns_book4.isna().sum()
```

```python
# 누락되지 않은 값을 확일할 때는 notna() 메서드 사용
ns_book4.notna().sum()
```

>>## 누락된 값으로 표시하기: None 과 np.nan

```python
# ns_book4 데이터프레임의 '도서권수' 열에 있는 첫 번째 행 값을 None으로 바꾼 후 '도서권수'열에 isna() 메서드를 적용해 누락된 값을 세어보기

ns_book4.loc[0, '도서권수'] = None
ns_book4['도서권수'].isna().sum()
```

```python
ns_book4.head(2)
# NaN을 실수로 저장하기 때문에 도서권수가 정수에서 실수형으로 변경됨.
```

```python
# '도서권수' 첫 번째 행 원래대로 바꾸기, 데이터 타입도 원래대로 변경

ns_book4.loc[0, '도서권수'] = 1
ns_book4 = ns_book4.astype({'도서권수': 'int32', '대출건수' : 'int32'}) # 매개변수를 딕셔너리 형식으로 전달
ns_book4.head(2)
```

```python
# 문자열 형식 열에 None 입력
import numpy as np
ns_book4.loc[0, '부가기호'] = np.nan
ns_book4.head(2)
```

>>## 누락된 값 바꾸기(1): loc,fillna() 메서드

```python
# 세트 ISBN 열의 누락된 값을 빈 문자열(' ') 로 변경
# 누락된 값을 가리키는 불리언 배열을 isna() 메서드로 만들어 loc 메서드에 전달

set_isbn_na_rows = ns_book4['세트 ISBN'].isna() # 누락된 값 불리언 배열 반환
ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
ns_book4['세트 ISBN'].isna().sum()
```

```python
# fillna()는 새로운 데이터프레임을 반환
ns_book4.fillna('없음').isna().sum()
```

```python
# 특정 열만 선택하여 NaN 바꾸기

ns_book4['부가기호'].fillna('없음').isna().sum()
```

```python
# '부가기호' 열의 NaN을 바꾸면서 전체 데이터프레임 반환

ns_book4.fillna({'부가기호':'없음'}).isna().sum() # 열 이름과 바꾸려는 값으로 이루어진 딕셔너리 전달
```

>>## 누락된 값 바꾸기(2): `replace()`
- replace(원래 값, 새로운 값) : 바꾸려는 값 하나일 때
- replace([원래 값1, 원래값2],[새로운 값1, 새로운 값2]) : 여러개일 때
- replace({열 이름: 원래 값}, 새로운 값) : 열 마다 다른 값으로 바꿀 때

```python
ns_book4.replace(np.nan, '없음').isna().sum()
```

```python
# 바꾸려는 값 여러개
ns_book4.replace([np.nan, '2021'],['없음',21]).head(2)
```

```python
# 딕셔너리 형식으로도 전달 가능
ns_book4.replace({np.nan:'없음', '2021':21}).head(2)
```

```python
# 열 마다 다른 값으로 바꿀 때

ns_book4.replace({'부가기호':np.nan},'없음').head(2)
```

```python
ns_book4.replace({'부가기호': {np.nan: '없음'}, '발행년도': {'2021': '21'}}).head(2)
```

>>## 숫차 찾기: \d
- 정규표현식 숫자 표현: \d
- 연도 표현: \d\d\d\d
- 뒤에 두자리 묶기: \d\d(\d\d)


```python
# 정규표현식 사용시 regex 매개변수 사용

ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]

# r 문자는 파이썬에서 정규 표현식을 다른 문자열과 구분하기 위해 접두사처럼 붙임
```

```python
# 정규표현식 반복시 중괄호를 사용하여 개수지정 가능

ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex = True)[100:102]
```

>>## 문자 찾기: 마침표(.)
- 어떤 문자에도 대응하는 정규표현식은 (.)이다.
- *를 사용하여 0개 이상 반복된다고 표시할 수 있다.
- 괄호()를 일반 문자로 인식하게 하려면 역슬래시를 앞에 붙여 쓴다. `\(\)`

```python
ns_book4.replace({'저자': {r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)': r'\1\2'}, '발행년도':{r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

># 잘못된 값 바꾸기

```python
# '발행년도' 열을 int32로 바꾸기
ns_book4.astype({'발행년도': 'int32'})
```

```python
# '1988.' 연도를 변환할 수 없어 오류가 발생.

ns_book4['발행년도'].str.contains('1988').sum() # '발행년도' 열에 '1988'이 포함된 행의 개수 세기
```

```python
# '발행년도' 열에서 숫자가 아닌 문자를 포함하는 모든 행 찾아보기
# na 매개변수 True로 지정하여 누락된 행을 True로 표시

invalid_number = ns_book4['발행년도'].str.contains('\D', na=True)
print(invalid_number.sum()) # 숫자 이외의 문자가 들어간 행의 개수를 출력
ns_book4[invalid_number].head()
```

```python
# 정규 표현식을 사용하여 연도 앞과 뒤에 있는 문자를 제외

ns_book5 = ns_book4.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex = True)
ns_book5[invalid_number].head()
```

```python
# 다시 숫자 이외의 문자가 들어간 행의 개수와 데이터를 확인

unknown_year = ns_book5['발행년도'].str.contains('\D', na = True)
print(unknown_year.sum())
ns_book5[unknown_year].head()
```

```python
# 변환되지 않은 값은 NaN이거나 네 자리 숫자가 아닌 값들이다.
# 임의로 -1로 바꾼 다음 astype() 메서드의 '발행년도'열의 데이터 타입을 정수형인 int32로 변환한다.

ns_book5.loc[unknown_year, '발행년도'] = '-1'
ns_book5 = ns_book5.astype({'발행년도':'int32'})
```

```python
# 연도 중 아주 큰 값이나 작은 값이 들어있는 경우를 확인
# 연도가 4000년이 넘는 경우 확인해보기

ns_book5['발행년도'].gt(4000).sum() # 연도가 4000년이 넘는 행의 전체 개수 확인
```

```python
# 위와 동일
(ns_book5['발행년도']>4000).sum()
```

```python
# 4000년이 넘는 연도에서 2333을 빼서 서기로 바꾼 다음 연도가 4000년이 넘는 도서가 있는지 확인

dangun_yy_rows = ns_book5['발행년도'].gt(4000) # 4000보다 큰 값을 찾는다.
ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows, '발행년도'] - 2333 # 찾은 값에서 2333을 빼서 서기로 바꿈

dangun_year = ns_book5['발행년도'].gt(4000) # 다시 4000보다 큰 값 찾기
print(dangun_year.sum())
ns_book5[dangun_year].head()
```

```python
# 연도가 비정상적으로 높은 도서는 13권, 모두 -1로 표시하기

ns_book5.loc[dangun_year, '발행년도'] = -1
```

```python
# 마지막으로 연도가 작은 값도 확인 / 0보다 크고 1900년도 이전의 도서 찾아보기

old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900) # lt 는 1900보다 작은 값 찾기

ns_book5[old_books]
```

```python
ns_book5.loc[old_books, '발행년도'] = -1
ns_book5['발행년도'].eq(-1).sum() # -1과 같은 값이 몇 개인지 센다.
```

># 누락된 정보 채우기

```python
# 발행년도가 -1 즉, 잘못 저장되어 있거나 알 수 없는 경우 채워줘야 함.
# '도서명', '저자', '출판사' 열에 누락된 값이 있거나 '발행년도'열이 -1인 행의 개수 확인

na_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna() | ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)

print(na_rows.sum())

ns_book5[na_rows].head()
```

```python
# 뷰티풀수프를 활용해 값 채우기

import requests
from bs4 import BeautifulSoup

# Yes24 에서 ISBN으로 검색한 결과 페이지에서 개발자 도구로 도서명 태그 정보 확인
# <a> 태그에 원하는 정보가 들어 있음을 확인할 수 있음

# 도서명을 가져오는 함수 작성

def get_book_title(isbn):
    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'

    # URL에 ISBN을 넣어 HTML 가져옴
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser') # HTML 파싱

    # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옴.
    title = soup.find('a', attrs={'class':'gd_name'}).get_text()

    return title

```

```python
# '골목의 시간을 그리다'책의 ISBN으로 테스트

get_book_title(9791191266054)
```

```python
# 마찬가지로 다른 정보들도 추출하여 반환
# 저자는 두 명 이상일 수 있기 때문에 find_all() 메서드를 사용해 저자를 담은 <a> 태그를 모두 추출

import re # re 모듈의 findall() 사용하면 정규식에 매칭되는 모든 문자열을 찾아 리스트로 반환

def get_book_info(row):
    title = row['도서명']
    author = row['저자']
    pub = row['출판사']
    year = row['발행년도']

    # Yes24 도서 검색 페이지 URL
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'

    # URL에 ISBN을 넣어 HTML 가져옴
    r = requests.get(url.format(row['ISBN']))
    soup = BeautifulSoup(r.text, 'html.parser') # HTML 파싱

    try:
        if pd.isna(title): # 도서명 비어 있는 경우
            # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져온다.
            title = soup.find('a', attrs={'class':'gd_name'}).get_text()

    except AttributeError:
        pass

    try:
        if pd.isna(author):
            # 클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴
            authors = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
            author_list = [auth.get_text() for auth in authors]
            author = ','.join(author_list)

    except AttributeError:
        pass


    try:
        if pd.isna(pub):
            # 클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴

            pub = soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()

    except AttributeError:
        pass


    try:
        if year == -1:
            # 클래스 이름이 'info_date'인 <span> 태그 아래 텍스트를 가져옴
            year_str = soup.find('span', attrs = {'class':'info_date'}).get_text()

            # 정규 표현식으로 찾은 값 중에 첫 번째 것만 사용
            year = re.findall(r'\d{4}', year_str)[0]

    except AttributeError:
        pass

    return title, author,pub,year
```

```python
# apply() 메서드는 기본적으로 반환된 값을 하나의 튜플로 만듬. result_type 매개변수를 'expand'로 지정하여 반환된 값을 각기 다른 열로 만들 수 있다.

updated_sample = ns_book5[na_rows].head(2).apply(get_book_info, axis=1, result_type = 'expand')
updated_sample
```

```python
# 모든 행에 적용할 수 있지만 시간 너무 오래걸림. 미리 만들어 놓은 데이터 프레임 다운

gdown.download('https://bit.ly/3UJZiHw', 'ns_book5_update.csv', quiet=False)
ns_book5_update = pd.read_csv('ns_book5_update.csv', index_col=0)
ns_book5_update.head()
```

```python
# ns_book5 데이터프레임을 ns_book5_update 데이터프레임 데이터로 업데이트 후 누락된 행 확인

ns_book5.update(ns_book5_update)
na_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna()| ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)
print(na_rows.sum())
```

```python
# 마지막으로 누락된 값을 가진 행을 삭제하여 분석 대상에서 제외

ns_book5 = ns_book5.astype({'발행년도': 'int32'})
ns_book6 = ns_book5.dropna(subset=['도서명','저자','출판사'])
ns_book6 = ns_book6[ns_book6['발행년도'] != -1]
ns_book6.head()
```
